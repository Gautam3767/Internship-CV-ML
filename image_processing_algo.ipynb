```python
from torchvision.transforms import v2
import torchvision.transforms as transforms
import cv2
import matplotlib.pyplot as plt
import torch
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('lena_clr.tiff', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('lena_clr.tiff')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(img)
```

```python
cv2_imshow(img2)
```

```python
th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(th1[1])
```

CANNY Edge Detection


```python
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img_rgb,-1,kernel)
plt.subplot(121),plt.imshow(img_rgb),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv2.blur(img_rgb2,(5,5))
```

```python
plt.subplot(121),plt.imshow(img_rgb2),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Gaussian Blur
blur2 = cv2.GaussianBlur(img_rgb2,(5,5),0)
plt.subplot(121),plt.imshow(img_rgb2),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur2),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Median Blurred
median = cv2.medianBlur(img_rgb2,5)
plt.subplot(121),plt.imshow(img_rgb2),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
